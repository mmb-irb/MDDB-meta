# Protein-nucleic acid interaction analysis
***
In this example we intend to analyze the interaction between the ***SARS-CoV-2 polymerase complex*** and a fragment of  ***RNA***.
We want to find and visualize the aminoacid with the highest ***electrostatic and Van der Waals energies*** and thus the one which is carrying the interaction.

For this test we will use Polymerase **atomistic MD trajectories** already loaded in the **[MDDB](https://irb.mddbr.eu) database** from different groups.

This workflow is based on the MDDB database REST API: https://irb.mddbr.eu/api/rest/docs/

## Complex-RNA interaction analysis

We will find which aminoacids in the polymerase complex are the most critical in the RNA interaction.<br />


```python
# Set a function to get the specific data energies we are interesed in
def get_energies_data (accession_or_id : str) -> dict:

    # Request the analysis
    analysis_request = API_BASE + 'projects/' + accession_or_id + '/analyses/energies'
    with urllib.request.urlopen(analysis_request) as response:
        parsed_response = json.loads(response.read().decode("utf-8"))
    analysis_data = parsed_response['data']

    # Get energies data for the specific interaction we are intereseted in: the complex-rna interaction
    interaction_data = next(interaction for interaction in analysis_data if 'complex–rna' in interaction['name'].lower() or 'complex-rna' in interaction['name'].lower())
    # Get specifically data for residues in the polymerase
    polymerase_data = interaction_data['agent1']

    # For this example, we are going to focus in the overall sum of energies
    energies_data = polymerase_data['both']
    
    return energies_data

# Set a function to get interaction residue indices from a specific simulation
def get_interaction_residue_indices (accession_or_id : str) -> list:
    
    # Use the simulation id to request the simulation interactions data
    interactions_request = API_BASE + 'projects/' + accession_or_id + '/analyses/interactions'
    with urllib.request.urlopen(interactions_request) as response:
        parsed_response = json.loads(response.read().decode("utf-8"))
    interactions = parsed_response
    
    # Get data for the specific interaction we are intereseted in: the complex-rna interaction
    interaction_data = next(interaction for interaction in interactions if 'complex–rna' in interaction['name'].lower() or 'complex-rna' in interaction['name'].lower())
    
    # We are interested in the polymerase reisude indices
    return interaction_data['residue_indices_1']

# Set a function to get the topology from a specific simulation
def get_topology (accession_or_id : str) -> dict:
    topology_request = API_BASE + 'projects/' + accession_or_id + '/topology'
    with urllib.request.urlopen(topology_request) as response:
        parsed_response = json.loads(response.read().decode("utf-8"))
    topology = parsed_response
    return topology

# Group energy values by residue
energies_per_residue = {}

# Iterate over the different projects
for project in projects:
    # Get the accession
    accession = project['accession']
    # Request the energies data
    energies_data = get_energies_data(accession)
    # Request the interactions data to know residue indices of the the energies data
    residue_indices = get_interaction_residue_indices(accession)
    # Request the topology to be able to parse residue indices to residue reference numeration
    # This allows us to group energy values by residue along the different simulations, even if their original numeration does not match
    topology = get_topology(accession)
    residue_reference_numbers = topology['residue_reference_numbers']
    # Now iterate over the different energy values and residue indices at the same time
    for energy, residue_index in zip(energies_data, residue_indices):
        # Parse the residue index to its reference numeration
        residue_reference_number = residue_reference_numbers[residue_index]
        # If the parsing returns null it may be due to a missalignment beacuse there is a mutation
        # Skip the value
        if residue_reference_number == None:
            continue
        # Add the current energy value to the energies per residue account
        current_energies = energies_per_residue.get(residue_reference_number, None)
        if current_energies:
            current_energies.append(energy)
        else:
            energies_per_residue[residue_reference_number] = [energy]
# Sort the energies per residue values according to residue numbers
energies_per_residue = { k: energies_per_residue[k] for k in sorted(energies_per_residue) }
# Now for each residue calculate the mean of its energy values
energy_means = [ mean(energies) for energies in energies_per_residue.values() ]
residue_numbers = list(energies_per_residue.keys())
# Add None values in residue gaps, in case there is any, for the graph to properly show the gaps
# Otherwise gaps would be connected by a stright line from the last existing value to the next existing value
i = 0
while residue_numbers[i] != residue_numbers[-1]:
    current_residue = residue_numbers[i]
    next_residue = residue_numbers[i+1]
    if next_residue != current_residue +1:
        energy_means.insert(i+1, None)
        residue_numbers.insert(i+1, current_residue +1)
        i += 1
    i += 1

# Display a plot with the mean energy values
fig = go.Figure({
    "data": [
        go.Scatter(
            x = residue_numbers,
            y = energy_means,
            connectgaps=False
        )
    ],
    "layout": go.Layout(
        title='Average electorstatic + Van der Waals energies per residue',
        xaxis={'title': 'Residues'},
        yaxis={'title': 'Electrostatic + VdW energy (kcal/mol)'},
    )
})
fig.show()
```

In the previous graph we can see 3 separated regions which belong to the 3 differen proteins in the polymerase complex.<br />
From left to right, they are the NSP8 cofactor, the NSP7 cofactor and the polymerase (NSP12).<br />
<br />
We can see some highly negative peaks. The highest peak belongs to residue 4892.<br />
Now we will pick one of the projects and download the structure and small subset of frames along the whole trajectory to visualize this residue and its dynamics.

```python
# Select one of the projects
sample_project = projects[0]
project_accession = sample_project['accession']

# Get the structure (pdb format)
structure_request = API_BASE + 'projects/' + project_accession + '/files/structure'
with urllib.request.urlopen(structure_request) as response:
    response_data = response.read().decode("utf-8")
structure_data = response_data

# Find out the total number of snapshots in the trajectory
total_frames = sample_project['metadata']['SNAPSHOTS']
print(f'Number of frames: {total_frames}')

# Calculate the frame step we need to get 10 frames along the trajectory
sample_frames = 10
frame_step = floor(total_frames / sample_frames)
print(f'Frame step: {frame_step}')

# The frame query format is <first frame>:<last frame>:<frame step>
frame_query = f'frames={1}:{total_frames}:{frame_step}'

# Get 25 frames along the trajectory (xtc format)
trajectory_request = API_BASE + 'projects/' + project_accession + '/files/trajectory?format=xtc&' + frame_query
with urllib.request.urlopen(trajectory_request) as response:
    response_data = response.read()
trajectory_data = response_data

# Write both structure and trajectory data to files
structure_filename = 'structure.pdb'
with open(structure_filename, 'w') as file:
    file.write(structure_data)
trajectory_filename = 'trajectory.xtc'
with open(trajectory_filename, 'wb') as file:
    file.write(trajectory_data)
print('Files written to ' + os.getcwd())
print(f' {structure_request} -> {structure_filename}')
print(f' {trajectory_request} -> {trajectory_filename}')
```

Now display the trajectory

```python
# Find residue 4892 in the topology
target_residue_reference_number = 4892
topology = get_topology(project_accession)
target_residue_index = topology['residue_reference_numbers'].index(target_residue_reference_number)
target_residue_number = topology['residue_numbers'][target_residue_index]
target_selection = str(target_residue_number)

# Show trajectory
view = nglview.show_simpletraj(nglview.SimpletrajTrajectory(trajectory_filename, structure_filename), gui=True)
view.add_representation(repr_type='ball+stick', selection=target_selection) # Highlight our target residue
view._remote_call('setSize', target='Widget', args=['800px','600px'])
view

# Focus in our target residue
code = """
var stage = this.stage;
var clist_len = stage.compList.length;
var i = 0;
var s = [];
for(i = 0; i <= clist_len; i++){
    if(stage.compList[i] != undefined && stage.compList[i].structure != undefined) {        
       s.push(stage.compList[i])
    }
}
"""
code += f's[ 0 ].autoView("{target_selection}")'
view._execute_js_code(code)
view
```
