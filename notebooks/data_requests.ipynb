# Extracting relevant information from atomistic Molecular Dynamics trajectories using the MDDB API
***
This **use case** aims to present several useful examples of how to **extract, play, store, and visualize** information from the **atomistic MD trajectories** stored in the **[MDDB](https://irb.mddbr.eu) database**. 


**INFOs** depending on the examples presented.


Although the **examples** are presented **step by step** with associated information, it is extremely advisable to previously spend some time reading documentation about **Molecular Dynamics simulations** and the generated **trajectories and metadata**, to get familiar with the terms used, especially for newcomers to the field. 

This workflow is based on the MDDB database REST API: https://irb.mddbr.eu/api/rest/docs/

### Importing required libraries

```python
import json, urllib, os
from math import floor

import ipywidgets as w
import plotly.graph_objs as go

import nglview
import simpletraj
```

***
## Input parameters
**Input parameters** which may be used to find simulations:
 - **unit**: Family of simulations.

Alternatively, the simulation could be **selected from the list of simulations** available in the **MDDB database** (See following cells) 
***

```python
# Set the API URL
API_BASE = "http://irb.mddbr.eu/api/rest/current"

# Get unit options
simulations_request = API_BASE + '/projects/options?projection=metadata.CV19_UNIT'
with urllib.request.urlopen(simulations_request) as response:
    response_data = json.loads(response.read().decode("utf-8"))
options = response_data['metadata.CV19_UNIT']

# Generating HTML table with options to choose
table = w.HTML("""
    <table style="width: 100%">
        <thead>
            <tr>
                <th>Family</th>
                <th>Number of simulations</th>
            </tr>
        </thead>
        <tbody>{}</tbody>
    </table>
""".format(
    ''.join([
        "<tr><td style='font-weight: bold; color: darkgreen; padding-right:10px;'>{}</td><td style='font-weight: bold; padding-right:10px;'>{}</td></tr>".format(
            name, count,
        )
        for name, count in options.items()
    ]))
)

display(table)

# Set dropdown menu to manually select another unit option
dropdown = w.Dropdown(
    options=list(options.keys()),
    description="Project:",
)
display(dropdown)

# Start with a particular family of simulations which may be changed by hand
unit = "RBD-ACE2"
def handle_change(change):
    global unit
    unit = change.get("new")
    print("unit selected: {}".format(unit))

dropdown.observe(handle_change, names="value")
```

Get a few simulations from this family to make an overview (10 simulations, by default)

```python
# Contact the MDDB database using the REST API
simulations_request = API_BASE + '/projects?query={"metadata.CV19_UNIT":"' + unit + '"}'
with urllib.request.urlopen(simulations_request) as response:
    response_data = json.loads(response.read().decode("utf-8"))
projects = response_data['projects']

# For each simulation (project) in the initial query, show its FAIR id, name and authors
for project in projects:
    print('\nProject ' + project['accession'])
    print('  - Name: ' + project['metadata']['NAME'])
    print('  - Authors: ' + ', '.join(project['metadata']['AUTHORS']))
```

## Getting information from a particular MD simulation

Using **REST API** from the **MDDB database** to extract all information stored for a particular **simulation**. See full **schemas** [here](https://irb.mddbr.eu/api/rest/docs/).

***

### Printing all metadata available

```python
sample_project = projects[0]
project_accession = sample_project['accession']
print("Displaying metadata for project " + project_accession)

# Getting info from a particular project_id
simulation_request = API_BASE + '/projects/' + project_accession
with urllib.request.urlopen(simulation_request) as response:
    response_data = json.loads(response.read().decode("utf-8"))

# Storing formatted info as JavaScript Object Notation (Json)
project = response_data
metadata = project['metadata']

# Showing info in Json format
print(json.dumps(metadata, indent=2))
```

```python
# Getting list of analysis available from project data
print("Simulation Analyses List:")
analyses = project["analyses"]

print(json.dumps(analyses,indent=2))
```

## Getting trajectory data from a particular MD simulation

Using **REST API** from the **MDDB database** to extract all or a subset of the trajectory stored for a particular **simulation**. See full **schemas** [here](https://irb.mddbr.eu/api/rest/docs/).

***

### Extracting trajectory data

Use the API to download the structure and a subset of the trajectory (a few frames along the whole trajectory)

```python
# Get the structure (pdb format)
structure_request = API_BASE + '/projects/' + project_accession + '/files/structure'
with urllib.request.urlopen(structure_request) as response:
    response_data = response.read().decode("utf-8")
structure_data = response_data

# Find out the total number of snapshots in the trajectory
total_frames = project['metadata']['SNAPSHOTS']
print(f'Number of frames: {total_frames}')

# Calculate the frame step we need to get 25 frames along the trajectory
sample_frames = 25
frame_step = floor(total_frames / sample_frames)
print(f'Frame step: {frame_step}' + str())

# The frame query format is <first frame>:<last frame>:<frame step>
frame_query = f'frames={1}:{total_frames}:{frame_step}'

# Get 25 frames along the trajectory (xtc format)
trajectory_request = API_BASE + '/projects/' + project_accession + '/files/trajectory?format=xtc&' + frame_query
with urllib.request.urlopen(trajectory_request) as response:
    response_data = response.read()
trajectory_data = response_data

# Write both structure and trajectory data to files
structure_filename = 'structure.pdb'
with open(structure_filename, 'w') as file:
    file.write(structure_data)
trajectory_filename = 'trajectory.xtc'
with open(trajectory_filename, 'wb') as file:
    file.write(trajectory_data)
print('Files written to ' + os.getcwd())
print(f'    {structure_request} -> {structure_filename}')
print(f'    {trajectory_request} -> {trajectory_filename}')
```

### Visualizing trajectory data

Use a molecular view (NGL) to display the trajectory

```python
# Show trajectory
view = nglview.show_simpletraj(nglview.SimpletrajTrajectory(trajectory_filename, structure_filename), gui=True)
view._remote_call('setSize', target='Widget', args=['800px','600px'])
view
```

## Getting analysis data from a particular MD simulation study

Using **REST API** from the **MDDB database** to extract data information from analyes performed for a particular **simulation**. See full **schemas** [here](https://irb.mddbr.eu/api/rest/docs/).
***

### Extracting analysis data

Use the API to download the RMSD analysis data

```python
# Get the RMSD analysis data
# Getting info from a particular project accession
simulations_request = API_BASE + '/projects/' + project_accession + '/analyses/rmsds'
with urllib.request.urlopen(simulations_request) as response:
    response_data = json.loads(response.read().decode("utf-8"))
rmsd_data = response_data
```

### Visualizing analysis data

Show a plot of the RMSD values along the trajectory

```python
# Mine the RMSD data from the analysis
sample_rmsd = next(rmsd for rmsd in rmsd_data['data'] if rmsd['group'] == 'protein' and rmsd['reference'] == 'firstframe')
rmsd_values = sample_rmsd['values']
rmsd_frames = [ i * rmsd_data['step'] for i in range(len(rmsd_values)) ]

# Set the plot
fig = go.Figure({
    "data": [ go.Scatter(x = rmsd_frames, y = rmsd_values) ],
    "layout": go.Layout(
        title = 'Protein RMSD per frame (against the first trajectory frame)',
        xaxis = {'title': 'Frames'},
        yaxis = {'title': 'RMSD (Å)'}
    )
})
fig.show()
```
