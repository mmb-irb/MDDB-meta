# Mining project data
***
This **tutorial** shows an example of how to mine data from a specific project in the **[MDDB](https://irb.mddbr.eu) database**. The project is the main entry in the database and it contains mostly metadata, but it is also an index of additional data which may be requested as well.

Although the **examples** are presented **step by step** with associated information, it is extremely advisable to previously spend some time reading documentation about **Molecular Dynamics (MD)** and the generated **trajectories and metadata**, to get familiar with the terms used, especially for newcomers to the field. 

This workflow is based on the MDDB database **[REST API](https://irb.mddbr.eu/api/rest/docs/)**:

#### Import required libraries

```python
import json, urllib
from os.path import exists
import plotly.graph_objs as go
```

#### Set some constants

```python
API_BASE_URL = "http://irb.mddbr.eu/api/rest/current"
```

#### Set a couple of functions to query the REST API

```python
# Set a function to call the API
def query_api (url : str) -> dict:
    # Parse the URL in case it contains any HTTP control characters
    # Replace white spaces by the corresponding percent notation character
    parsed_url = url.replace(" ", "%20")
    with urllib.request.urlopen(parsed_url) as response:
        return json.loads(response.read().decode("utf-8"))
# Set a function to call the API
def download_file_api (url : str, filename : str):
    # Parse the URL in case it contains any HTTP control characters
    # Replace white spaces by the corresponding percent notation character
    parsed_url = url.replace(" ", "%20")
    urllib.request.urlretrieve(url, filename)
```

## The project data scheme

A project provides topology and most simulation parameters to one ore more MDs. This allows to group MDs with identical or very similar configurations (e.g. replicas) as a single entry in the database.

The **request** to the REST API to retrieve a specific project is done by adding the **project accession or identifer** after the projects endpoint. In addition, to request an specific MD in the selected project, both the accession and the identifier may be followed by a dot and the **number of the MD** to be returned. This number is 1-based. In this example we use project with accession MCV1900193. This project has 3 MDs and the reference MD is the first one. This means that requesting both MCV1900193 and MCV1900193.1 returns the same data while MCV1900193.2 and MCV1900193.3 are different from the reference.

The **response** of the REST API when requesting a project contains the following fields:
- accession and identifier: Both unique labels of the project. A project has an accession only if it is published while all projects have an identifier. Accession is human firendly while the identifier is database friendly.
- published: true if the project is published. Note that non-published projects are hidden and they may be not reachable depending on the API base URL you are using.
- metadata: project name, authors, simulation parameters, structural features, etc. Most metadata is associated to the project and thus it does not change among different MDs. 
- chains: A list with names of protein chains analyzed by the protein functional analysis. Chains do not change between MDs.
- mds: Available MDs inside the project. Note that this field is returned only when there are multiple MDs.
- mdref: Reference MD is the MD to be returned by default, when no MD number is provided in the request (this is explained below in more detail). Note that mdref is a 0-based index while the accession MD number is 1-based.
- mdNumber: The current MD number. This is 1-based.
- analyses: A list of available analyses corresponding to the requested MD.
- files: A list of available files corresponding to the requested MD.
- other: depending on the project there may be additional fields

Below we request the second MD in the project with accession MCV1900193 to display the whole response as an example.<br />
Note that you can also do the query and see the response in [your browser](http://irb.mddbr.eu/api/rest/current/projects/MCV1900193.2).

```python
# Set the URL for the specific project
specific_project_query = API_BASE_URL + '/projects/MCV1900193.2'
print('We query the API at ' + specific_project_query)

# Query the API
response = query_api(specific_project_query)
# Print the response by pieces to make it prettier
for field_name, field_value in response.items():
    if type(field_value) == list:
        print(field_name + ':')
        for value in field_value:
            print('  - ' + str(value))
    elif type(field_value) == dict:
        print(field_name + ':')
        for key, value in field_value.items():
            print('  - ' + key + ': ' + str(value))
    else:
        print(field_name + ': ' + str(field_value))
```

## Requesting analysis data

Available analyses are listed in the project.<br />
This list of analysis names is also returned when querying the **'analyses' endpoint**.

```python
# Set the URL for the specific project analyses
analyses_query = specific_project_query + '/analyses'
print('We query the API at ' + analyses_query)

# Query the API
response = query_api(analyses_query)
# Print the available analyses
print('Available analyses: ')
for analysis in response:
    print('- ' + analysis)
```

For this tutorial we want to obtain the RMSD data and plot a graph.<br />
This graph should be very similar to the one we see in the [web page](https://irb.mddbr.eu/#/id/MCV1900193.2/rmsds).

In order to request a given analysis we must provide its name after the analyses endpoint as it is done in the following example:

```python
# Set the URL for the specific analysis
rmsd_query = analyses_query + '/rmsds'
print('We query the API at ' + rmsd_query)

# Query the API
rmsd_data = query_api(rmsd_query)
# Note that data is not displayed here since it would overflow
```

Finally, mine the RMSD data and plot the graph

```python
# Mine the RMSD data from the analysis
sample_rmsd = next(rmsd for rmsd in rmsd_data['data'] if rmsd['group'] == 'protein' and rmsd['reference'] == 'firstframe')
rmsd_values = sample_rmsd['values']
rmsd_frames = [ i * rmsd_data['step'] for i in range(len(rmsd_values)) ]

# Set the plot
fig = go.Figure({
    "data": [ go.Scatter(x = rmsd_frames, y = rmsd_values) ],
    "layout": go.Layout(
        title = 'Protein RMSD per frame (against the first trajectory frame)',
        xaxis = {'title': 'Frames'},
        yaxis = {'title': 'RMSD (Å)'}
    )
})
fig.show()
```

## Downloading a file


***

In a similar way, files are listed in the project and the list of filenames may be requested using the **'files' endpoint**.

```python
# Set the URL for the specific project files
files_query = specific_project_query + '/files'
print('We query the API at ' + files_query)

# Query the API
response = query_api(files_query)
# Print the available files
print('Available files: ')
for filename in response:
    print('- ' + filename)
```

An especific file may be retrieved also by adding its name the to files endpoint.<br />
In this example we want to download the 'topology.psf' file, which is the original topology file.

```python
# Set the URL for the specific file
target_filename = 'topology.psf'
specific_file_query = files_query + '/' + target_filename
print('We query the API at ' + specific_file_query)

# Query the API
download_file_api(specific_file_query, target_filename)
if exists(target_filename):
    print('Target file has been downloaded succesfully')
```
